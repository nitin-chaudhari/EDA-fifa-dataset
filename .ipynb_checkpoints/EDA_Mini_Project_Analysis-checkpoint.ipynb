# Data Preprocessing

### 1. Reading the data

```python
fifa_data = pd.read_csv('fifa.csv')
fifa_variable_information = pd.read_csv('fifa_ variable_information.csv')
```

### 2. Dropping unnecessary columns

```python
fifa_data.drop(['Photo','Club Logo'],inplace=True,axis=1)
```

```python
fifa_data.shape
```

### 3. Converting datatypes for following columns.

Value,Wage,Joined,Contract Valid Untill,Height,Weight,Release Clause

```python
fifa_data['Value'].dtype
```

```python
for i in range(len(fifa_data['Value'])):
    if '.' in fifa_data['Value'].iloc[i]:
        fifa_data['Value'].iloc[i] = fifa_data['Value'].iloc[i].replace('.','')
        fifa_data['Value'].iloc[i] = fifa_data['Value'].iloc[i].replace('M','00000')
    else:
        fifa_data['Value'].iloc[i] = fifa_data['Value'].iloc[i].replace('M','000000')       
```

```python
fifa_data['Value'] = fifa_data['Value'].str.replace('K','000')
fifa_data['Value'] = fifa_data['Value'].str.replace('€','')

fifa_data.Value = fifa_data['Value'].astype('float64')
```

```python
fifa_data['Value'].dtypes
```

```python
fifa_data['Value'].head()
```

```python
fifa_data['Wage'].dtype
```

```python
fifa_data['Wage'] = fifa_data['Wage'].str.replace('K','000')
fifa_data['Wage'] = fifa_data['Wage'].str.replace('€','')

fifa_data['Wage'] = fifa_data['Wage'].astype('float64') 
```

```python
fifa_data['Wage'].dtype
```

```python
fifa_data['Wage'].head()
```

```python
fifa_data['Joined'].dtype
```

```python
fifa_data['Joined'] = pd.to_datetime(fifa_data['Joined'])
fifa_data['Joined'] = fifa_data['Joined'].dt.year
fifa_data['Joined'] = np.nan_to_num(fifa_data['Joined']).astype(int)
fifa_data['Joined'] = fifa_data['Joined'].astype('int64')
```

```python
fifa_data['Joined'].dtype
```

```python
fifa_data['Joined']
```

```python
fifa_data['Contract Valid Until'].dtype
```

```python
fifa_data['Contract Valid Until'] = fifa_data['Contract Valid Until'].astype('datetime64')
```

```python
fifa_data['Contract Valid Until'].head()
```

```python
fifa_data['Height'].dtype
```

```python
fifa_data['Height'] = fifa_data['Height'].str.replace("'",'.').astype('float64')
```

```python
fifa_data['Height'].dtype
```

```python
fifa_data['Height'].head()
```

```python
fifa_data['Weight'].dtype
```

```python
fifa_data['Weight'] = fifa_data['Weight'].str.replace('lbs','').astype('float64')
```

```python
fifa_data['Weight'].dtype
```

```python
fifa_data['Weight'].head()
```

```python
fifa_data['Release Clause'].dtype
```

```python
fifa_data['Release Clause'].head()
```

```python
fifa_data['Release Clause'] = fifa_data['Release Clause'].replace(np.nan,'0')
```

```python
for i in range(len(fifa_data['Release Clause'])):
    if '.' in fifa_data['Release Clause'].iloc[i]:
        fifa_data['Release Clause'].iloc[i] = fifa_data['Release Clause'].iloc[i].replace('.','')
        fifa_data['Release Clause'].iloc[i] = fifa_data['Release Clause'].iloc[i].replace('M','00000')
    else:
        fifa_data['Release Clause'].iloc[i] = fifa_data['Release Clause'].iloc[i].replace('M','000000')  
```

```python
fifa_data['Release Clause'] = fifa_data['Release Clause'].str.replace('K','000')
fifa_data['Release Clause'] = fifa_data['Release Clause'].str.replace('€','')

fifa_data['Release Clause'] = fifa_data['Release Clause'].replace('0',np.nan).astype('float64')
```

```python
fifa_data['Release Clause'].dtype
```

```python
fifa_data['Release Clause']
```

#### a)	Are there any players appearing in more than one Table. Please point out such players.

Making list having tables of top 5 players in each unique position.

```python
list_of_tables = [RF_POS,ST_POS,LW_POS,GK_POS,RCM_POS,LF_POS,RS_POS,RCB_POS,LCM_POS,
                  CB_POS,LDM_POS,CAM_POS,CDM_POS,LS_POS,LCB_POS,RM_POS,LM_POS,
                  LB_POS,RDM_POS,RW_POS,CM_POS,RB_POS,RAM_POS,CF_POS,LAM_POS,RWB_POS,LWB_POS]
```

```python
len(list_of_tables)
```

```python
### Here we are checking if any player appearing in more than one table or not.

more_than_in_one = [] ##### List of players more than in one table.
for i in range(len(list_of_tables)):
    table_1 = list_of_tables[i]
    for j in range(i+1,len(list_of_tables)):
        table_2 = list_of_tables[j]
        for k in range(5):
            if table_1['Name'].iloc[k] in table_2['Name']:
                more_than_in_one.append(table_1['Name'].iloc[k])
print("Player more than in one table are :",len(more_than_in_one))
```

So there are no players appearing more than one table.

#### b)	What is the average wage one can expect to pay for the top 5 in every position? 

Making list having each unique position name.

```python
pos_name = list(fifa_data['Position'].unique())
```

```python
len(pos_name)
```

```python
### Here we are calculating the average wage for top 5 player in every position.

avg_wage_evry_pos = dict() ##### Dictionary having position name as key and average wage as value.
for i in range(27):
    avg_wage_evry_pos.update({pos_name[i]:list_of_tables[i]['Wage'].mean()})
print("The average wage one can expect to pay for top 5 in every position :\n\n",avg_wage_evry_pos)  
```

Making DataFrame having each unique position with respective their average wage.

```python
avg_wage_data = pd.DataFrame({'Position':avg_wage_evry_pos.keys(),'Avg_Wage':avg_wage_evry_pos.values()})
avg_wage_data
```

Plotting graph for each unique positions with average wage.

```python
plt.figure(figsize=(16,7))
sns.barplot(x=avg_wage_data['Position'],y=avg_wage_data['Avg_Wage'])
plt.title("Plot for Average Wage for top 5 player in every position")
plt.xlabel('Position',fontsize='15')
plt.ylabel('Average Wage',fontsize='15')
plt.show()
```
